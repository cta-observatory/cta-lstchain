This is an example notebook to show how to use pandas for quick data exploration and results visualition based on a results hdf5 file.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:98% !important; }</style>"))
```

```python
gammas_table = pd.read_hdf('gamma_events_point_tiny.h5')
protons_table = pd.read_hdf('proton_events_tiny.h5')
```

```python
gammas_table.describe()
```

```python
protons_table.describe()
```

```python
hist_opt = {'figsize':(18,18), 'density':True, 'alpha':0.6, 'bins':80}
axes = gammas_table.hist(label='diffuse gammas', **hist_opt);
axes = protons_table.hist(ax=axes.ravel()[:min(len(protons_table.columns),len(gammas_table.columns))],
                          label='protons', **hist_opt);
```

## Correlation matrices

```python
import seaborn as sns
```

```python
# Gammas
f, ax = plt.subplots(figsize=(10, 8))
features = gammas_table.columns
corr = gammas_table[features].corr()
sns.heatmap(corr, 
            mask=np.zeros_like(corr, dtype=np.bool),
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, 
            ax=ax);
```

```python
# Protons
f, ax = plt.subplots(figsize=(10, 8))
features = protons_table.columns
corr = protons_table[features].corr()
sns.heatmap(corr, 
            mask=np.zeros_like(corr, dtype=np.bool),
            cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, 
            ax=ax);
```
